1.How many males and females plot the count of gender?

```python
import plotly.express as px
fig = px.histogram(sales_df,
             x='Gender',
             y='User_ID',
             title='User_ID vs Gender',
             color='Gender',
             color_discrete_sequence=["black", "blue"]) # specify colors here
fig.show()
```

3.count of each age group with individual count of age group wrt gender?




```python
age_gender_counts = sales_df.groupby(['Age Group', 'Gender'])['User_ID'].count().reset_index(name='Count')
print(age_gender_counts)
```

```python
import plotly.express as px

fig = px.bar(age_gender_counts,
             x='Age Group',
             y='Count',
             color='Gender',
             title='Count of Each Age Group by Gender',
             barmode='group')
fig.show()
```

4.plot the total amt spent by each age group?

5.plot the graph that depicts total number of orders from top 10 states

```python
state_orders = sales_df.groupby('State')['Orders'].sum().sort_values(ascending=False).head(10).reset_index(name='Orders')
print(state_orders)
```

```python
import plotly.express as px
fig = px.histogram(state_orders,
             x='State',
             y='Orders',
             title='Orders vs State',
             color='State',
             color_discrete_sequence=["red", "green"]) # specify colors here
fig.show()
```

6. Total amt of top 10 states?

```python
state_amount = sales_df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(10).reset_index(name='Amount')
print(state_amount)
```

```python
import plotly.express as px
fig = px.histogram(state_amount,
             x='State',
             y='Amount',
             title='Amount vs State',
             color='State',
             color_discrete_sequence=["red", "green"]) # specify colors here
fig.show()
```

7.plot me a graph that shows comparison between no of married and unmarried people?

```python
import plotly.express as px

# Replace values in the 'Marital_Status' column before passing it to px.histogram
sales_df['Marital_Status'] = sales_df['Marital_Status'].replace({0: 'unmarried', 1: 'married'})

fig = px.histogram(sales_df,
             x='Marital_Status',  # Now use the modified column
             y='User_ID',
             title='User_ID vs Marital_Status',
             color='Marital_Status',
             color_discrete_sequence=["red", "green"])
fig.show()
```

8.plot me amt spent by each male and each femlae wrt to marital status?

```python
gender_marital_amount = sales_df.groupby(['Gender', 'Marital_Status'])['Amount'].sum().reset_index(name='Amount')
gender_marital_amount
```

```python
gender_marital_amount = sales_df.groupby(['Gender', 'Marital_Status'])['Amount'].sum().sort_values(ascending=True).reset_index(name='Amount').replace({0:'unmarried', 1:'married'})

import plotly.express as px
fig = px.bar(gender_marital_amount,
             x='Gender',
             y='Amount',
             color='Marital_Status',
             title='Amount Spent by Gender and Marital Status',
             barmode='group')
fig.show()
```

10. Plot the amt spent by each occupation in descending order?

```python
occupation_amount = sales_df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False).reset_index(name='Amount')
print(occupation_amount)
```

11.plot a graph that provides statistics on the intrest of customers wrt to product category?

```python
import plotly.express as px
fig = px.histogram(sales_df,
             x='Product_Category',
             y='User_ID',
             title='User_ID vs Product_Category',
             color='Product_Category')
fig.show()
```

12.i would like to know the budget spent on each product category in descending order?

```python
amt_product= sales_df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).reset_index(name='Amount')
print(amt_product)
```

```python
import plotly.express as px
fig = px.bar(amt_product,
             x='Product_Category',
             y='Amount',
             title='Amount vs Product_Category',
             color='Product_Category')
fig.show()
```

13. Provide me statical analysis of each product based upon the percentage of orders that has been completed?

```python
orders_product= sales_df.groupby('Product_Category')['Orders'].sum().sort_values(ascending=False).reset_index(name='Orders')
print(orders_product)
```

```python
import plotly.express as px
fig = px.histogram(orders_product,
             x='Product_Category',
             y='Orders',
             title='Orders vs Product_Category',
             color='Product_Category')
fig.show()
```
